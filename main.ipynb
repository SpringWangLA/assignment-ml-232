```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = np.array([
  [1, 1, 1],
  [1, 2, 1],
  [1, 3, 1],
  [2, 1, 1],
  [2, 2, 1],
  [2, 3, 1],
  [2, 3.5, 1],
  [2.5, 2, 1],
  [3.5, 1, 1],
  [3.5, 2, 1],
  [3.5, 3, 2],
  [3.5, 4, 2],
  [4.5, 1, 2],
  [4.5, 2, 2],
  [4.5, 3, 2],
  [5, 4, 2],
  [5, 5, 2],
  [6, 3, 2],
  [6, 4, 2],
  [6, 5, 2]
])

features = data[:, :2]  # Select features (first 2 columns)
labels = data[:, -1]    # Extract class labels (last column)
features
```

```python
import matplotlib.pyplot as plt


features = data[:, :-1]  # Select features (all columns except the last - label)
labels = data[:, -1]    # Extract class labels (last column)

# Scatter plot with color-coding based on labels
scatter = plt.scatter(features[:, 0], features[:, 1], c=labels)

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Features with Class Labels')

# Show the plot
plt.legend(handles=scatter.legend_elements()[0], labels=['Class 1', 'Class 2'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data_frame = pd.DataFrame(data, columns=["Feature 1", "Feature 2", "Class Label"])
print(data_frame)
correlations = data_frame.corr(method='pearson')  # Calculate Pearson correlation
plt.matshow(correlations)  # Create heatmap

# Add labels and title
plt.xticks(range(len(correlations.columns)), correlations.columns, rotation=45)
plt.yticks(range(len(correlations.columns)), correlations.columns)
plt.colorbar(label='Correlation Coefficient')
plt.title('Correlation between Features and Class Label')

# Display the plot
plt.show()
```

```python

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
print(scaled_features)
print('Mean of dataset:', scaler.mean_)
print('Standard Deviation of dataset:', scaler.scale_)
```

```python
# Apply PCA for dimensionality reduction (e.g., reduce to 2 components)
n_components = 1
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_features)
print('========================Thông số sau khi PCA=======================')
print('explained_variance_ratio_: ',pca.explained_variance_ratio_)
print('components_: ', pca.components_)
print('========================Dữ liệu sau khi PCA=======================')
if n_components == 2:
  print(pd.DataFrame(pca.components_,columns=['Feature 1', 'Feature 2'],index = ['PC-1','PC-2']))
  print(pd.DataFrame(pca_data, columns=['Feature 1', 'Feature 2']))
elif n_components == 1:
  print(pd.DataFrame(pca.components_,columns=['Feature 1', 'Feature 2'],index = ['PC-1']))
  print(pd.DataFrame(pca_data, columns=['Feature 1']))
```

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

LRM = LogisticRegression()
DTC = DecisionTreeClassifier()
RFC = RandomForestClassifier()
KNC = KNeighborsClassifier()
NBC = GaussianNB()

SKF = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 10)
for i, (train_index, test_index) in enumerate(SKF.split(features, labels)):
  print(f"Fold {i}:")
  print(f"  Train: index={train_index}")
  print(f"  Test:  index={test_index}")

print('============Kết quả một số mô hình ML cơ bản với bộ dữ liệu sau PCA============')
print(f'LogisticRegression : {round(cross_val_score(LRM,pca_data,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'DecisionTreeClassifier : {round(cross_val_score(DTC,pca_data,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'RandomForestClassifier : {round(cross_val_score(RFC,pca_data,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'KNeighborsClassifier : {round(cross_val_score(KNC,pca_data,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'GaussianNB : {round(cross_val_score(NBC,pca_data,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')

print('============Kết quả một số mô hình ML cơ bản với bộ dữ liệu trước PCA============')
print(f'LogisticRegression : {round(cross_val_score(LRM,features,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'DecisionTreeClassifier : {round(cross_val_score(DTC,features,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'RandomForestClassifier : {round(cross_val_score(RFC,features,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'KNeighborsClassifier : {round(cross_val_score(KNC,features,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
print(f'GaussianNB : {round(cross_val_score(NBC,features,labels,cv=SKF,scoring="accuracy").mean()*100,2)}%')
```
